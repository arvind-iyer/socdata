```python
def daily_complaints(year=2015):
    start = datetime(year, 1, 1)
#     dc = {i:[] for i in range(1,13)} 
    dc = []
    next_day = lambda starttime : len(between_dates(start, start+one_day))
    if start.year % 4 == 0:
        days_in_year = 366
    else:
        days_in_year = 365
            
    for i in range(days_in_year):
        dc.append(next_day(start))
#         dc[start.month].append(next_day(start))
        start = start + one_day
    return dc
# len(between_dates(start, start+one_hour))


def hourly_complaints(hours=24):
    start = datetime(2015, 1, 1)
    if hours > 24:
        hours = 24
    hc = []
    next_hour = lambda starttime : len(between_hours(start, start + one_hour))
    for i in range(hours):
        hc.append(next_hour(start))
        start = start + one_hour
    return hc
```
